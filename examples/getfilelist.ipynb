## カレントディレクトリのファイル一覧を表示

ファイル一覧には%FileクラスのFileSetクエリを使用します。

```arbitrary
set stmt=##class(%SQL.Statement).%New()
set sc=stmt.%PrepareClassQuery("%File","FileSet")
if sc'=1 { write $SYSTEM.Status.ErrorText(sc),! quit }
```

クエリを実行し、結果セット result を取得します。

```arbitrary
set result=stmt.%Execute()
```

%Displayを使用すると、すべての結果を表示します。

```arbitrary
do result.%Display()
```

ファイル名のみを取得する場合、while ループを使用します

```arbitrary
while result.%Next() { write result.%Get("Name"),! }
```
